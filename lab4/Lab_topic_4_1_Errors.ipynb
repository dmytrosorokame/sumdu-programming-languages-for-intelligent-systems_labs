## 3. raise
**Trigger an exception manually**

Частина 3: raise

ЗАВДАННЯ:
Створи свою функцію, яка викидає ValueError, якщо число більше 100.

### Sometime we can create an error on purpose

```python
i = 33;
if i !=3:
    raise OSError()
```

```python
try:
    q = 1 / 0
    print(q)
except ZeroDivisionError:
    print("Check the division")
    q = 1
```

```python
def check_value(n):
    if n > 100:
        raise ValueError(f"Value {n} exceeds the allowed maximum of 100")
    return n

try:
    print(check_value(42))
    print(check_value(150))
except ValueError as e:
    print("Caught:", e)
```

## 4. assert
**Conditional trigger an exception in the code**

Частина 4: assert
Продемонструвати assert як засіб перевірки

Пояснити, коли доцільно використовувати (при розробці, не у продакшні)

ЗАВДАННЯ:
Реалізуй функцію, яка приймає число і перевіряє, що воно додатне, з допомогою assert.

```python
def f(x):
    assert x < 0, 'x must be negative'
    return x**2
```

```python
f(1)
```

```python
f(-1)
```

```python
def check_positive(n):
    assert n > 0, f"Expected a positive number, got {n}"
    return n

print(check_positive(10))
print(check_positive(0.5))

try:
    check_positive(-3)
except AssertionError as e:
    print("AssertionError:", e)
```

## Conclusions

In this lab we looked at 5 ways to handle errors in Python:

- `try/except` catches errors and lets the program keep running instead of crashing. You can catch specific errors like `OSError` or `ZeroDivisionError` to react differently.

- `try/except/finally` - `finally` always executes, even if there was an error. Good for cleanup, like closing files.

- `raise` throws errors manually when some condition is not met. We used it to raise `ValueError` when a number was too large.

- `assert` is a quick way to check conditions during development. Raises `AssertionError` if the check fails, but can be disabled with `-O` flag so not for production.

- `with/as` opens files and closes them automatically when the block ends. Much simpler than calling `f.close()` every time.
